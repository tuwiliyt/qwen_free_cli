## Important: Configure Your Token

Before running the agent, make sure you have edited `credentials.json` with your actual token.

You can edit it directly in Google Drive or use the cell below to check:

```python
# Check if credentials.json is configured
import json
try:
    with open('credentials.json', 'r') as f:
        creds = json.load(f)
    token = creds.get('sessions', [{}])[0].get('qwen_credentials', {}).get('access_token', '')
    if token and token != 'PASTE_YOUR_TOKEN_HERE':
        print('✅ Credentials configured correctly')
    else:
        print('❌ Please edit credentials.json with your actual token')
except FileNotFoundError:
    print('❌ credentials.json not found')
```
